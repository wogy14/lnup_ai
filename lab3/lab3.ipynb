- Узбережжя 4 970 км²

```python
def numberToBinary(number):
    assert isinstance(number, float)

    result = '0|' if number >= 0 else '1|'
    nums = str(abs(number)).split(".")

    result += bin(int(nums[0])).lstrip("0b")

    if len(nums) > 1:
        result += '.'
        dec = float('0.' + nums[1])
        for i in range(10):
            dec *= 2
            result += str(int(dec))
            if dec == 1.0:
                break
            dec -= int(dec)

    return result
```
